```python
# Removing Redundant measurements
temp = []
for i in df['size']:

  if str(i).strip().split('/')[0] != 'nan':
    if len(str(i).split('/')) == 1:
      temp.append(str(i).strip())
    else:
      temp.append(str(i).strip().split('/')[1].strip())
  else:
      temp.append(str(i).strip())
df['size'] = temp
```

```python
df1 = pd.read_csv('/content/reviews_0-250_info.csv', engine='python', on_bad_lines='skip')
df2 = pd.read_csv('/content/reviews_250-500_info.csv', engine='python', on_bad_lines='skip')
df3 = pd.read_csv('/content/reviews_500-750_info.csv', engine='python', on_bad_lines='skip')
df4 = pd.read_csv('/content/reviews_750-1250_info.csv', engine='python', on_bad_lines='skip')
df5 = pd.read_csv('/content/reviews_1250-end_info.csv', engine='python', on_bad_lines='skip')
reviews_merged = pd.concat([df1,df2,df3,df4,df5])
reviews_merged.to_csv('merged_reviews.csv',index=False)
```

```python
reviews_merged.drop(['helpfulness','total_feedback_count','total_neg_feedback_count','total_pos_feedback_count','submission_time','skin_tone','eye_color','skin_type','hair_color','product_name','brand_name'],inplace=True,axis=1)
```

```python
import spacy

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Function to extract adjectives from text, while handling non-string data
def extract_adjectives(text):
    if isinstance(text, str):  # Check if the entry is a string
        doc = nlp(text)  # Process the text with spaCy
        adjectives = [token.text for token in doc if token.pos_ == 'ADJ']  # Extract adjectives
        return adjectives
    else:
        return []  # Return an empty list for non-string entries

# Apply the function to the 'review_text' column
reviews_merged['adjectives'] = reviews_merged['review_title'].apply(extract_adjectives)
```
